```python
from os import listdir
import pandas as pd
```

```python
files = listdir("../chart")

dfs = {}
for f in files:
    year = int(f.split("_")[0])
    month = f.split("_")[1].split(".")[0]

    if dfs.get(year) is None:
        dfs[year] = {}
        
    dfs[year][month] = pd.read_csv("../chart/" + f)
```

```python
pays = []
nationalite = []
naissance = []

for k1 in dfs.keys():
    for k2 in dfs[k1].keys():
        pays += dfs[k1][k2]["Pays d'origine"].tolist()
        nationalite += dfs[k1][k2]["Nationalité"].tolist()
        naissance += dfs[k1][k2]["Naissance"].tolist()
```

```python
for p in set(nationalite):
    if not any(x.lower() in p.lower()  for x in ["Français", "Française", "Belge", "Canadien", "Canandienne", "France", "Algérienne", "Marocaine", "Tunisienne", "française", "Franco", "Canada"]):
        print(p)
```

```python
for k, p in enumerate(set(pays)):
    if not any(x.lower() in p.lower()  for x in ["france", "belge"]):
        print(f"{k : 3} - {p}")
```

```python
for n in set(naissance):
    print(n.split(" ")[:3])
```

```python
for k1 in dfs.keys():
    for k2 in dfs[k1].keys():
        if dfs[k1][k2]["Pays d'origine"].str.contains("Canada").any():
            print(f"{k1} - {k2}")
```
